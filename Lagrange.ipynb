```python
import matplotlib.pyplot as plt

# Define the points
xs = [-2, 1, 3, 7]
ys = [5, 7, 11, 34]

#Figura con tama√±o de 15 y 10
plt.figure(figsize=(15, 10))

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Grafico de puntos ")

# Add a grid for better readability
plt.grid(True)

# Display the plot
plt.show()
```

```python
def f(x: float) -> float:
    L0 = (
        (x - xs[1])
        * (x - xs[2])
        * (x - xs[3])
        / ((xs[0] - xs[1]) * (xs[0] - xs[2]) * (xs[0] - xs[3]))
    )
    L1 = (
        (x - xs[0])
        * (x - xs[2])
        * (x - xs[3])
        / ((xs[1] - xs[0]) * (xs[1] - xs[2]) * (xs[1] - xs[3]))
    )
    L2 = (
        (x - xs[0])
        * (x - xs[1])
        * (x - xs[3])
        / ((xs[2] - xs[0]) * (xs[2] - xs[1]) * (xs[2] - xs[3]))
    )
    L3 = (
        (x - xs[0])
        * (x - xs[1])
        * (x - xs[2])
        / ((xs[3] - xs[0]) * (xs[3] - xs[1]) * (xs[3] - xs[2]))
    )
    return L0 * ys[0] + L1 * ys[1] + L2 * ys[2] + L3 * ys[3]
```

```python
import numpy as np

# Generate x values
x_values = np.linspace(min(xs), max(xs), 100)

# Compute y values using the function f
y_values = [f(x) for x in x_values]

# Plot the function
plt.plot(x_values, y_values, label="f(x)")
plt.scatter(xs, ys,color="purple", label="Grafica")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lagrange interpolation")
plt.legend()
plt.show()
```
